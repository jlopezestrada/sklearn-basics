# Linear Regression
---

Linear Regression is a statistical method used to model the relationship between a dependent variable and one or more independent ones by fitting a linear equation on the observed data.

## Implementation using Scikit-Learn
---

To implement the linear regression algorithm, we'll start by obtaining some sample data. For this example, we'll utilize the datasets provided by Scikit-Learn.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
```

### Data Exploration of the dataset: California Housing
---

```python
data = fetch_california_housing()

X = data.data
y = data.target
features = data.feature_names

df = pd.DataFrame(X, columns=features)
df['MedHouseVal'] = y

df.hist(bins=50, figsize=(15, 10))
plt.show()
```

### Implementation of Linear Regression model
---

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"MSE: {mse}")
```

```python
plt.figure(figsize=(10, 6)) 
plt.plot(y_test, label='Actual Values', color='blue', marker='o', linestyle='-', markersize=5)  # Actual values
plt.plot(y_pred, label='Predicted Values', color='red', marker='x', linestyle='--', markersize=5)  # Predicted values
plt.title('Comparison of Actual and Predicted Values')
plt.xlabel('Sample Index')
plt.ylabel('Median House Value')
plt.legend()
plt.grid(True)
plt.show()
```
