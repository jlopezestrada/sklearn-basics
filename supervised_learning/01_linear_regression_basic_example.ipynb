# Linear Regression
---

Linear Regression is a statistical method used to model the relationship between a dependent variable and one or more independent ones by fitting a linear equation on the observed data.

## Implementation using Scikit-Learn
---

To implement the linear regression algorithm, we'll start by obtaining some sample data. For this example, we'll utilize a very simple dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Basic Dataset
---
For this example dataset, we are going to simple linear function $y = 4 + 3x$. Gaussian noise will be added to simulate real-world conditions.

```python
np.random.seed(502) # Seed to maintain the values for the example

x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
y_true = 4 + 3 * x

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data points with noise')
plt.plot(x, y_true, color='red', label='Target')
plt.legend()
plt.show()
```

### Implementation of Linear Regression model
---

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"MSE: {mse}")

plt.figure(figsize=(10, 6))
plt.title("Target vs Predicted Linear Regression")
plt.scatter(x, y, color='blue', label='Data points with noise')
plt.plot(x, y_true, color='red', label='Target')
plt.plot(X_test, y_pred, color='green', label='Predicted')
plt.legend()
plt.show()
```
